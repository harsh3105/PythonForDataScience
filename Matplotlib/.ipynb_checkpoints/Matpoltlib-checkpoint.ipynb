```python
import matplotlib.pyplot as plt
```

```python
 %matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y=x**2
```

```python
x
```

```python
y
```

```python
#functional
```

```python
plt.plot(x,y)
plt.xlabel('x label')
plt.ylabel('y label')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(y,x)
```

```python
#OO
```

```python
fig= plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig,axes= plt.subplots(nrows=1,ncols=2)
plt.tight_layout()

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
a=axes[0].plot(x,y)
```

```python
#Figure and sizes and DPI
```

```python
fig = plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig("pic.png",dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="x**2")
ax.plot(x,x**3,label="x**3")

ax.legend(loc=0)
```
